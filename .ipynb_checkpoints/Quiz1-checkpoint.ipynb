```python
# სვეტის ინდექსირება ხდება დეილი რანკის მიხედვით
df = pd.read_csv('data/universal_top_spotify_songs.csv', index_col='daily_rank') 
# schema_df = pd.read_csv('data/universal_top_spotify_songs_schema.csv', index_col='Column' )
```

```python
# გვაქვს 10 სვეტი და 100 რიგი
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 100)
```

```python
df.head(10)
```

```python
# ფილტრავს ტოპ 10 სიმღერას, პოპულარობის დონე მაღალი უნდა იყოს 90ზე
filt_popularity = df['popularity']>90
df.loc[filt_popularity,['popularity', 'name']].head(10)
```

```python
# df['name'].head(10)
```

```python
# ალაგებს ქვეყნისა და პუპულარობის მიხედვით ხოლო შედეგს გვიჩვენებს სახელსა და სიმღერის პოპულარობის მაჩვენებელს
df.sort_values(by=['country', 'popularity'], ascending=[True, False], inplace=True)
df[['name', 'popularity']].head(10)
```

```python
# popularity სვეტის მიხედვით ითვლის სტანდარტულ გადახრას
df['popularity'].std()
```

```python
# popularity სვეტის მიხედვით ითვლის პოპულარობისსაშუალო რიცხვს
df['popularity'].mean()
```

```python
# energy სვეტის მიხედვით ითვლის მედიანას
df['energy'].median()
```

```python
# energy სვეტში პოულობს მინიმალურ რიცხვს
df['energy'].min()
```

```python
# energy სვეტში პოულობს მაქსიმალურ რიცხვს
df['energy'].max()
```

```python
# პოპულარობის მიხედვით პირველ 10 სიმღერას გვიჩვენებს
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots(figsize=(15, 10))
name = df['name'].head(10)
popularity = df['popularity'].head(10)

bar_colors = ['g']

ax.bar(name, popularity, color=bar_colors, width=0.5)

ax.set_xlabel('name', fontsize=25)
ax.set_ylabel('popularity', fontsize=25)
ax.set_title('song names and popularity')


plt.show()
```

```python
#არტისტების tempoს პროცენტულობას ასახავს, პირველი 10 პარამეტრია გამოტანილი
import matplotlib.pyplot as plt
labels = df['artists'].head(10)
sizes = df["tempo"].head(10)
fig, ax = plt.subplots()

ax.pie(sizes, labels=labels, autopct='%.0f%%',
       textprops={'size': 'smaller'}, radius=1)
plt.show()
```
